{ *** 1.  COLLECT THE COMPETITION DATA FROM THE API ENDPOINTS *** }

```python

import requests

url = "https://api.sportradar.com/tennis/trial/v3/en/competitions.json?api_key=3cx83hgyhyf1YOaeIdPIxYu10bw5H0JSANgpnVSI"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
cat_comp_data=response.json()
```

```python
# data extracted use for loop :

category_id = []
category_name = []
for value in cat_comp_data["competitions"]:
    category_id.append(value['category']['id'])
    category_name.append(value['category']['name'])

data = {'category_id': category_id, 'category_name': category_name}

category_df = pd.DataFrame(data)

category_df
```

```python
category_df.nunique()
```

```python
category_df.drop_duplicates(inplace=True)
category_df
```

```python
category_df.to_csv("category_cleaned.csv",index=False)
```

```python
# data extracted use for loop :

competition_id = []
competition_name = []
parent_id = []
competition_type = []
gender = []
category_id = []

for value in cat_comp_data["competitions"]:
    competition_id.append(value.get('id'))
    competition_name.append(value.get('name'))
    parent_id.append(value.get('parent_id'))
    competition_type.append(value.get('type'))
    gender.append(value.get('gender'))
    category_id.append(value['category']['id'])

data = {'competition_id': competition_id, 'competition_name': competition_name, 'parent_id': parent_id, 'type': competition_type, 'gender': gender, 'category_id': category_id}
competitions_df = pd.DataFrame(data)


competitions_df.fillna("NULLABLE",inplace=True)

competitions_df
```

```python
competitions_df.to_csv("competition_cleaned.csv",index=False)
```

{ *** 2. COLLECT THE COMPLEXES DATA FROM THE API ENDPOINTS *** }

```python
import requests

url = "https://api.sportradar.com/tennis/trial/v3/en/complexes.json?api_key=3cx83hgyhyf1YOaeIdPIxYu10bw5H0JSANgpnVSI"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
compl_data=response.json()
```

FETCH DATA FROM API AND CONVERTED INTO DATAFRAME FOR CATEGORY_DF

```python
# data extracted use for loop :

complex_data=[]

for complexes in compl_data["complexes"]:
     complex_info = {
          "complex_id": complexes.get("id"),
          "complex_name": complexes.get("name")
        
     }
     complex_data.append(complex_info)

complex_df = pd.DataFrame(complex_data)


complex_df
```

```python
complex_df.nunique()
```

```python
complex_df.drop_duplicates(inplace=True)
complex_df
```

```python
complex_df.to_csv("complex_cleaned.csv",index=False)
```

FETCH DATA FROM API AND CONVERTED INTO DATAFRAME FOR VENUES_DF


```python
# data extracted use for loop :

venue_data=[]
complexes = compl_data['complexes']

for complex_info in compl_data["complexes"]:
    for venue in complex_info.get('venues', []):
        venue_info = {
            "complex_id": complex_info['id'],
            
            "venue_id": venue['id'],
            "venue_name": venue['name'],
            "city_name": venue['city_name'],
            "country_name": venue['country_name'],
            "country_code": venue['country_code'],
            "timezone": venue['timezone']
        }
        venue_data.append(venue_info)

venues_df=pd.DataFrame(venue_data)
venues_df
```

```python
venues_df.to_csv("venues_cleaned.csv",index=False)
```

{ *** 3. COLLECT THE DOUBLES COMPETITOR RANKINGS DATA FROM THE API ENDPOINTS ***  }


```python
import requests

url = "https://api.sportradar.com/tennis/trial/v3/en/double_competitors_rankings.json?api_key=3cx83hgyhyf1YOaeIdPIxYu10bw5H0JSANgpnVSI"

headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)

print(response.text)
```

```python
double_competitor_data=response.json()
```

FETCH DATA FROM API AND CONVERTED INTO DATAFRAME FOR DOUBLE_COMPTITOR_DF


```python
# data extracted use for loop :

double_rank=[]
for value in double_competitor_data["rankings"]:
    for data in value.get( "competitor_rankings",[]):
        rank_data={
            "rank":data["rank"],
            "movement":data["movement"],
            "points":data["points"],
            "competitions_played":data["competitions_played"],
            "competitor_id":data["competitor"].get("id")
        }
        double_rank.append(rank_data)

double_competitor_df=pd.DataFrame(double_rank)        

double_competitor_df
```

```python
double_competitor_df.nunique()
```

```python
double_competitor_df.drop_duplicates(inplace=True)
double_competitor_df
```

```python
double_competitor_df.to_csv("double_competitor_cleaned.csv",index=False)
```

FETCH DATA FROM API AND CONVERTED INTO DATAFRAME FOR COMPTITORS_DF

```python
# data extracted use for loop :

competitors =[]
for value in double_competitor_data["rankings"]:
    for data in value.get( "competitor_rankings",[]):
        compet_data={
            "competitor_id":data["competitor"].get("id"),
            "name":data["competitor"].get("name"),
            "country":data["competitor"].get("country"),
            "country_code":data["competitor"].get("country_code"),
            "abbreviation":data["competitor"].get("abbreviation")
        }
        competitors.append(compet_data)

competitors_df=pd.DataFrame(competitors)

competitors_df
```

```python

competitors_df.fillna("NULL",inplace=True)
competitors_df
```

```python
competitors_df.to_csv("competitors_cleaned.csv",index=False)
```
