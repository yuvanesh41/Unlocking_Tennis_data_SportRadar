```python
#                                      COMPETITOR_RANKINGS_TABLE DATA STORE IN POSTGRESQL
import pandas as pd
import psycopg2

# Establish the database connection

connection = psycopg2.connect(
    host="localhost",
    user="postgres",
    password="vijay45",
    database="tennis_data"
)
cursor = connection.cursor()

# Import the CSV file
df = pd.read_csv(r'C:\Users\aswin\3D Objects\mdtm\double_competitor_cleaned.csv')
double_competitor_rankings = df.copy()

# Data Preprocessing

# Information about dataset
double_competitor_rankings.info()

# Drop duplicates based on the 'rank' column (update column name if needed)
double_competitor_rankings.drop_duplicates(subset=['rank'], inplace=True)

# Check for null values
print(double_competitor_rankings.isna().sum())

# Create table if it doesn't already exist
create_table_query = '''
CREATE TABLE IF NOT EXISTS competitor_rankings_table (
    rank_id SERIAL,
    rank INT NOT NULL,
    movement INT NOT NULL,
    points INT NOT NULL,
    competitions_played INT NOT NULL,
    competitor_id VARCHAR(50),
    PRIMARY KEY (rank_id),
    FOREIGN KEY (competitor_id) REFERENCES competitors_table(competitor_id)
);
'''
cursor.execute(create_table_query)
print('The table has been created (or already exists).')

# Prepare the insert query
insert_into = '''INSERT INTO competitor_rankings_table
                 (rank, movement, points, competitions_played, competitor_id)
                 VALUES (%s, %s, %s, %s, %s)'''

# Convert dataframe to list of tuples for batch insertion
data_to_insert = double_competitor_rankings[['rank', 'movement', 'points', 'competitions_played', 'competitor_id']].values.tolist()

# Insert data in bulk using executemany for better performance
try:
    cursor.executemany(insert_into, data_to_insert)
    connection.commit()
    print("Data has been uploaded successfully.")
except Exception as e:
    print(f"Error during bulk insert: {e}")


# Ensure the data has been stored
cursor.execute('SELECT COUNT(*) FROM competitor_rankings_table')
row_count = cursor.fetchone()[0]
print(f'The data has been uploaded. Total rows in table: {row_count}')
```
